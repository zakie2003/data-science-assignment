```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from imblearn.over_sampling import SMOTE
f=pd.read_csv("./toy_data.csv")
new_cols=pd.get_dummies(f["Type"])*1
f=f.drop(["Type"],axis=1)
f=pd.concat([new_cols,f],axis=1)
minmax=MinMaxScaler()
f[["Air temperature [K]","Process temperature [K]","Rotational speed [rpm]","Torque [Nm]","Tool wear [min]"]]=minmax.fit_transform(f[["Air temperature [K]","Process temperature [K]","Rotational speed [rpm]","Torque [Nm]","Tool wear [min]"]])
f=f.drop(["UDI","Product ID"],axis=1)
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
f['Failure Type']= label_encoder.fit_transform(f['Failure Type'])
res=f["Failure Type"]
f=f.drop(["Failure Type"],axis=1)
```

```python
f
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_train,x_test,y_train,y_test=train_test_split(f,res,test_size=0.2)
x_train,y_train=smote.fit_resample(x_train,y_train)
x_test,y_test=smote.fit_resample(x_test,y_test)
```

```python
from tensorflow import keras
def neural_net():
    md=keras.Sequential([
        keras.layers.Flatten(),
        keras.layers.Dense(8,input_shape=(1000,8),activation="relu"),
        keras.layers.Dense(50,activation="relu"),
        keras.layers.Dense(6,activation="sigmoid")

    ])
    md.compile(optimizer="adam",loss='sparse_categorical_crossentropy',metrics=["accuracy"])
    return md
```

```python
m=neural_net()
m.fit(x_train,y_train,epochs=10)
```

```python
m.evaluate(x_test,y_test)
```
