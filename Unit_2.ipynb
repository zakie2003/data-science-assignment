**Q1)Program to add Index Collum**

**Q2)Program to  represent categorical variable**

```python
import numpy as np
import pandas as pd
f=pd.read_csv("./Datasets/toy_data.csv")
new_cols=pd.get_dummies(f["Type"])*1
```

```python
f=f.drop(["Type"],axis=1)
new_f=pd.concat([new_cols,f],axis=1)
print(new_f)
```

**Q2.4)Program to get standardized value**

#### **Q2.6)Perform Binning On a dataset**

```python
from sklearn.cluster import KMeans
import seaborn as sb
import numpy as np
X = np.array([[1, 2], [1, 4], [1, 0],[1,3],[1,5],[1,1],
               [10, 2], [10, 4], [10, 0],[10,1],[10,5],[10,3]])
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(X)
label=kmeans.labels_

sb.scatterplot(x=X[:,0],y=X[:,1],hue=label)
```
