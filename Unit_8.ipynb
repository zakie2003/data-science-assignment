#### **Program for demostrating kmean clustering**

```python
from sklearn.cluster import KMeans
import seaborn as sb
import numpy as np
X = np.array([[1, 2], [1, 4], [1, 0],[1,3],[1,5],[1,1],
               [10, 2], [10, 4], [10, 0],[10,1],[10,5],[10,3]])
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(X)
label=kmeans.labels_

```

#### **Before applying Kmean Cluster**

```python
sb.scatterplot(x=X[:,0],y=X[:,1],color=["red"])
```

#### **After applying Kmean Cluster**

```python
sb.scatterplot(x=X[:,0],y=X[:,1],hue=label)
```
